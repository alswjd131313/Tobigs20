## 1. library와 framework의 차이 간단하게 서술하기

- 라이브러리는 공통으로 사용될 수 있는 특정한 기능들의 집합을 의미한다. 특정한 기능만을 수행하도록 제작된 코드를 개발자가 호출해 사용한다. 동작하는 완전한 프로그램이 아니라 특정한 부분 기능만을 수행하도록 제작된 컴파일되어 기계어의 형태로 존재하는 프로그램이다. 
- 프레임워크는 하나의 프로그램을 개발할 때 개발의 다양한 기능들을 통합한 공통적인 개발환경을 제공해주는 것을 말한다. 어떤 태그를 써야하는지, 어떤 함수를 추가로 작성해야하는지, 소스 파일을 어느 위치에 넣어야 하는지 등이 정해져있다. 개발에 필요한 화면 구현, DB연동, 개발환경들에 공통적인 부분을 제공해 개발시간과 리소스 비용을 절감해준다. 
- 프레임워크는 여러 기능을 가진 클래스와 라이브러리가 특정 결과물을 구현하고자 합쳐진 형태로 프레임워크는 라이브러리보다 상위 그룹이라고 볼 수 있다. 라이브러리는 단순한 클래스의 집합체이고, 프레임워크는 여러기능을 가진 클래스와 라이브러리가 합쳐진 형태이다. 

## 2. 딥러닝과 머신러닝의 관계 및 특징, 차이 간단하게 서술하기

인공지능의 분야 내에 머신러닝이 포함되고, 머신러닝 내에 딥러닝이 포함된다. 머신러닝은 컴퓨터 시스템이 데이터로부터 학습하고 패턴을 식별하고 예측을 수행할 수 있는 인공지능의 한 분야를 뜻하고, 딥러닝은 인공신경망 알고리즘을 이용한 머신러닝의 한 분야를 의미한다. 머신러닝은 예측 분석, 패턴 인식, 추천 시스템 등의 분야에 사용되고, 딥러닝은 이미지 및 음성 인식, 자연어 처리, 자율주행 등의 분야에 주로 사용된다. 
머신러닝의 경우 데이터 기반으로 알고리즘과 모델을 학습해 예측, 분류, 군집화 등 작업을 수행하고, 딥러닝은 다층 신경망을 사용한다. 또한 딥러닝은 사람이 패턴을 분석할 필요없이 분류에 사용할 데이터를 스스로 학습할 수 있는 반면 머신러닝은 분석하려는 학습 데이터를 수동으로 제공해 패턴을 추출하는 방법을 명령해줘야 컴퓨터가 스스로 데이터를 분석하고 수집한다는 차이가 있다.
